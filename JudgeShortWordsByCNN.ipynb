# **録音を開始した時間を0sにする**

```python
#timeデータの録音開始した秒数を0sにする関数
def TimesSetting0(timesSameVowelsDatas):
    adjustTimes = []
    #発話1回のデータを参照する
    for i in range(len(timesSameVowelsDatas)):
        #発話1回分のデータが録音を開始したときの秒数をstartTimeにいれる
        startTime = timesSameVowelsDatas[0]
        #adjustTimesにSetのそれぞれのtimeデータ-startTimeを入れる
        adjustTimes.append(timesSameVowelsDatas[i]-startTime)
    return adjustTimes
```

# **5000ms～8000msの間のデータを取り出す**

```python
#Setに入っている時間と気圧データの5000ms～8000msの間を取り出す関数
def CuttingData5000from8000(adjustTimes, presses):
    cuttingPresses5000from8000 = []
    cuttingTimes5000from8000 = []
    loop_range = min(len(adjustTimes), len(presses))

    for i in range(len(adjustTimes)):
        if (5000 <= adjustTimes[i]) and (adjustTimes[i] <= 8000):
            cuttingPresses5000from8000.append(presses[i])
            cuttingTimes5000from8000.append(adjustTimes[i])

    return cuttingPresses5000from8000, cuttingTimes5000from8000
```

# **cpu/gpuの選択とエポック回数について決める**

```python
class CNF():
    ngpu = 0 #-1はcpuを利用，gpuを使いたいときは0以上にする
    device = torch.device("cuda:0" if (torch.cuda.is_available() and ngpu > 0) else "cpu")
    epochs = 100 #学習回数
```

```python
import itertools

data = []
dataNum = []
for k in range(numDataInSetTry):
  data.append([])
  dataNum.append([])

#データをセットごとに5つに分ける
for i in range(numOfSetTry):
    for j in range(5):
        for k in range(numDataInSetTry):
            #Set1のとき
            if i==0:
              data[k] = [pressesAdjustAve[i][0][k], pressesAdjustAve[i][1][k], pressesAdjustAve[i][2][k], pressesAdjustAve[i][3][k], pressesAdjustAve[i][4][k]]
              dataNum[k].append([j])

trainX = []
trainy = []

#とりあえずは1つ目のデータをtestデータにする
for k in range(numDataInSetTry):
  testX = data[0]
  testy = dataNum[0]
  if k!=0:
    trainX.append(data[k])
    trainy.append(dataNum[k])

print("trainX↓")
print(np.array(trainX))
print("trainy↓")
print(np.array(list(itertools.chain.from_iterable(trainy))))
print("testX↓")
print(np.array([testX]))
print("testy↓")
print(np.array(testy))

class MyDataset(object):
  def __init__(self, X, y):
      self.X = X
      self.y = y

  def __getitem__(self, index:int):
      X, y = self.X[index], self.y[index]
      return X, y

  def __len__(self):  # Python2だと next(self) で定義
      return len(self.X)

#読み込んだデータをPyTorchの Dataset オブジェクトに変換し、reshape メソッドでデータの形状を調整し、CNN (畳み込みニューラルネットワーク) に入力できる形式にする
train_data = MyDataset(np.array(trainX).reshape(-1, 5, 90), np.array(list(itertools.chain.from_iterable(trainy))))
test_data = MyDataset(np.array([testX]).reshape(-1, 5, 90), np.array(testy))
```

# **CNNの処理class**

```python
class Net(nn.Module):
    def __init__(self, channel, kernel, h1, h2):
        super(Net, self).__init__() #初期化
        self.conv1 = ppe.nn.LazyConv1d(None, channel, kernel) #1次元畳み込み層を定義
        self.pool = nn.MaxPool1d(2, 2) #最大プーリング層を定義、入力データの次元の削減で使用、プーリングウィンドウのサイズとストライドを表す
        self.conv2 = ppe.nn.LazyConv1d(None, channel, kernel) #1次元畳み込み層を定義
        #全結合層を定義-------------------------
        self.fc1 = ppe.nn.LazyLinear(None, h1) #最初の2つの全結合層の出力ユニット数を表す
        self.fc2 = ppe.nn.LazyLinear(None, h2) #最初の2つの全結合層の出力ユニット数を表す
        self.fc3 = ppe.nn.LazyLinear(None, 6) #全結合層を定義
        #---------------------------------------

    def forward(self, x): #ニューラルネットワークへの入力データがどのように処理されるかの定義
        x = self.pool(F.relu(self.conv1(x))) #入力データを畳み込み層、ReLU活性化関数、そして最大プーリング層に通す
        x = self.pool(F.relu(self.conv2(x))) #上記と同様
        #view メソッドを使用してテンソルの形状を変更する
        x = x.view(x.size(0), -1)  # バッチサイズを維持し、残りを1次元に変換
        x = x.flatten(start_dim=1) #プーリング層の出力を平坦化する
        x = F.relu(self.fc1(x)) #平坦化されたデータを全結合層とReLU活性化関数に通す
        x = F.relu(self.fc2(x)) #上記と同様
        x = self.fc3(x) #データを最後の全結合層に通す
        return F.log_softmax(x, dim=1) #出力にログソフトマックス関数を適用し、結果を返す。ログソフトマックス関数は、出力を確率分布に変換するために使用する。
```

# **cnnを繰り返して、精度を向上させる関数**

```python
#ハイパーパラメータの最適化
#Optunaを使って、ハイパーパラメータを最適化する際に、モデルの性能を評価するための目的関数
def objective(trial):
    lr = trial.suggest_float('lr', 1e-5, 1e-2, log=True) #学習率（lr）を0.00001 から 0.01 までの範囲で、対数スケールで探索
    batch_size = trial.suggest_int('batch_size', 2, 4096) #一度に処理するデータサンプル数（batch_size）を2 から 4096 までの範囲で探索
    channel = trial.suggest_int('channel', 2, 128) #畳み込み層のチャネル数(channel)を2 から 128までの範囲を探索 。チャネル数は、畳み込みニューラルネットワーク (CNN) において、特徴マップの数を表します。特徴マップは、入力データから抽出された特徴を表現するもので、チャネル数を増やすことで、より多くの特徴を学習することができます。
    kernel = trial.suggest_int('kernel', 2, 20) #入力データに対して適用されるフィルター（kernel）を2 から 32 までの範囲で探索
    h1 = trial.suggest_int('h1', 2, 128) #ニューロン数（h1）を2 から 128 までの範囲で探索。ニューロン数は、その層が持つノードの数を表し、モデルの表現力に影響を与えます。
    h2 = trial.suggest_int('h2', 2, 128) #全結合層の2番目の隠れ層のニューロン数（h2）を2 から 128 までの範囲で探索。h1と同様。

    train_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size,
                                             shuffle=True)
    test_loader = torch.utils.data.DataLoader(test_data, batch_size=batch_size,
                                             shuffle=True)

    dataloaders_dict = {}
    dataloaders_dict["train"] = train_loader
    dataloaders_dict["test"] = test_loader

    model = Net(channel, kernel, h1, h2)
    dummy_input, _ = next(iter(train_loader))
    model(dummy_input.float())

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr = lr)

    accuracy_lst, accuracy_val_lst = [], []
    loss_lst, loss_val_lst = [], []
    model.to(CNF.device)
    for epoch in range(CNF.epochs):

        for phase in ["train", "test"]:
            if phase == "train":
                model.train()
            else:
                model.eval()

            loss_epoch = 0.0
            acc_epoch = 0.0

            if (epoch == 0) and (phase == "train"):
                continue

            for inputs, labels in dataloaders_dict[phase]:
                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == "train"):
                    labels = labels.to(CNF.device)
                    inputs = inputs.to(CNF.device)
                    outputs = model(inputs.float())

                    loss = criterion(outputs, labels.flatten())
                    _, preds = torch.max(outputs, 1)

                    if phase == "train":
                        # fisrt forward-backward pass
                        model.zero_grad()
                        loss.backward()
                        optimizer.step()

                    loss_epoch += loss.item() * inputs.size(0)
                    acc_epoch += torch.sum(preds == labels.data.flatten())

            loss_epoch = loss_epoch / len(dataloaders_dict[phase].dataset)
            acc_epoch = acc_epoch.double() / len(dataloaders_dict[phase].dataset)

            print(f"phase: {phase}",
                  f"epoch: {epoch}",
                  f"loss: {loss_epoch:.4f}",
                  f"accuracy: {acc_epoch:.4f}")
                 # f"accuracy: {acc_epoch}:.4f}")

            if phase == "train":
                accuracy_lst.append(acc_epoch)
                loss_lst.append(loss_epoch)
            else:
                accuracy_val_lst.append(acc_epoch)
                loss_val_lst.append(loss_epoch)
        #過学習防止のための早期終了
        min_idx = loss_val_lst.index(min(loss_val_lst))
        if epoch - min_idx > 5:
          """
          plt.plot(accuracy_val_lst, label="test")
          plt.plot(accuracy_lst, label="train")
          plt.xlabel("epoch")
          plt.ylabel("accuracy")
          plt.legend()
          plt.show()

          plt.plot(loss_val_lst, label="test")
          plt.plot(loss_lst, label="train")
          plt.xlabel("epoch")
          plt.ylabel("loss")
          plt.legend()
          plt.show()
          """
          break
    """
    plt.plot(accuracy_val_lst, label="test")
    plt.plot(accuracy_lst, label="train")
    plt.xlabel("epoch")
    plt.ylabel("accuracy")
    plt.legend()
    plt.show()

    plt.plot(loss_val_lst, label="test")
    plt.plot(loss_lst, label="train")
    plt.xlabel("epoch")
    plt.ylabel("loss")
    plt.legend()
    plt.show()
    """

    return max(accuracy_val_lst)
```
