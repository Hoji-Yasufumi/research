# **録音を開始した時間を0sにする**

```python
#timeデータの録音開始した秒数を0sにする関数
def TimesSetting0(timesSameVowelsDatas):
    adjustTimes = []
    #発話1回のデータを参照する
    for i in range(len(timesSameVowelsDatas)):
        #発話1回分のデータが録音を開始したときの秒数をstartTimeにいれる
        startTime = timesSameVowelsDatas[0]
        #adjustTimesにSetのそれぞれのtimeデータ-startTimeを入れる
        adjustTimes.append(timesSameVowelsDatas[i]-startTime)
    return adjustTimes
```

```python
timesAdjustSet0 = [] #調整した経過時間を入れておくための配列

#配列の作成----------------------------------
for i in range(numOfSetTry):
    timesAdjustSet0.append([])
    for j in range(5):
        timesAdjustSet0[i].append([])
        for k in range(numDataInSetTry):
            timesAdjustSet0[i][j].append([])
#---------------------------------------

for i in range(numOfSetTry):
    for j in range(5):
        for k in range(numDataInSetTry):
            timesAdjustSet0[i][j][k] = TimesSetting0(times[i][j][k])
```

# **5000ms～8000msの間のデータを取り出す**

```python
#Setに入っている時間と気圧データの5000ms～8000msの間を取り出す関数
def CuttingData5000from8000(adjustTimes, presses):
    cuttingPresses5000from8000 = []
    cuttingTimes5000from8000 = []
    loop_range = min(len(adjustTimes), len(presses))

    for i in range(len(adjustTimes)):
        if (5000 <= adjustTimes[i]) and (adjustTimes[i] <= 8000):
            cuttingPresses5000from8000.append(presses[i])
            cuttingTimes5000from8000.append(adjustTimes[i])

    return cuttingPresses5000from8000, cuttingTimes5000from8000
```

```python
cuttingPresses5000from8000 = [] #5000～8000msの間の気圧を入れておく
cuttingTimes5000from8000 = [] #5000～8000msの間の時間を入れておく

#配列の作成-----------------------------------------------------
for i in range(numOfSetTry):
    cuttingPresses5000from8000.append([])
    cuttingTimes5000from8000.append([])
    for j in range(5):
        cuttingPresses5000from8000[i].append([])
        cuttingTimes5000from8000[i].append([])
        for k in range(numDataInSetTry):
            cuttingPresses5000from8000[i][j].append([])
            cuttingTimes5000from8000[i][j].append([])
#----------------------------------------------------------------

#データを5000ms～8000msに切り取る---------------------
for i in range(numOfSetTry):
    for j in range(5):
        for k in range(numDataInSetTry):
            cuttingPresses5000from8000[i][j][k], cuttingTimes5000from8000[i][j][k] = CuttingData5000from8000(timesAdjustSet0[i][j][k], presses[i][j][k])
#-------------------------------------------------------
```

```python
def AveSdMaxMin(pressAdjustAve, minDistanceForDTW):
    avePress = 0 #平均
    sdPress = 0 #標準偏差
    maxPress = 0 #最大値
    minPress = 0 #最小値
    #maxPresses10 = [] #最大値から10個のデータ
    #minPresses10 = [] #最小値から10個のデータ
    aveMaxPress10 = 0 #最大値から10個のデータの平均
    aveMinPress10 = 0 #最小値から10個のデータの平均
    distanceOngaku = 0 #Ongakuと比較したときのdtw距離
    distanceSaisei = 0 #Saiseiと比較したときのdtw距離
    distanceTeisi = 0 #Teisiと比較したときのdtw距離
    distanceTugihe = 0 #Tugiheと比較したときのdtw距離
    distanceModoru = 0 #Modoruと比較したときのdtw距離

    #print(pressAdjustAve)

    if len(pressAdjustAve) >= 2:
        avePress = sta.mean(pressAdjustAve)
        sdPress = sta.stdev(pressAdjustAve)
        maxPress = max(pressAdjustAve)
        minPress = min(pressAdjustAve)
        maxPresses10 = sorted(pressAdjustAve, reverse=True)[:10]
        minPresses10 = sorted(pressAdjustAve)[:10]
        aveMaxPress10 = sta.mean(maxPresses10)
        aveMinPress10 = sta.mean(minPresses10)
        distanceOngaku = minDistanceForDTW[0][0]
        distanceSaisei = minDistanceForDTW[0][1]
        distanceTeisi = minDistanceForDTW[0][2]
        distanceTugihe = minDistanceForDTW[0][3]
        distanceModoru = minDistanceForDTW[0][4]


    else:
        avePress = 0
        sdPress = 0
        maxPress = 0
        minPress = 0
        aveMaxPress10 = 0
        aveMinPress10 = 0
        distanceOngaku = 0
        distanceSaisei = 0
        distanceTeisi = 0
        distanceTugihe = 0
        distanceModoru = 0

    return avePress,sdPress, maxPress, minPress, aveMaxPress10, aveMinPress10, distanceOngaku, distanceSaisei, distanceTeisi, distanceTugihe, distanceModoru
```

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# ==========================
# 1. データセットの準備
# ==========================
# 今回はサンプルとして、make_classificationで2クラス分類用の合成データを作成します
# n_samples: サンプル数
# n_features: 特徴量の次元
# n_informative: 分類に有効な特徴量の数
# random_state: 乱数の種
X, y = make_classification(n_samples=1000,
                           n_features=10,
                           n_informative=2,
                           n_redundant=0,
                           n_classes=2,
                           random_state=42)

print(X)
print(y)

# ==========================
# 2. StratifiedKFoldの準備
# ==========================
# k=5の交差検証を行う (5分割)
# shuffle=True で各エポックごとにデータをシャッフル
# random_state=42 で再現性のある分割
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# ==========================
# 3. 交差検証の実行
# ==========================
accuracies = []

for fold_idx, (train_index, test_index) in enumerate(skf.split(X, y), start=1):
    # トレーニングデータとテストデータを分割
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # 分割後のクラスの分布を確認 (任意)
    # np.bincountはクラスのカウントを返す
    print(f"Fold {fold_idx}")
    print("  Train クラス分布:", np.bincount(y_train))
    print("  Test  クラス分布:", np.bincount(y_test))

    # ==========================
    # 4. モデルの学習
    # ==========================
    model = LogisticRegression()
    model.fit(X_train, y_train)

    # テストデータで予測
    y_pred = model.predict(X_test)

    # ==========================
    # 5. 評価指標の計算
    # ==========================
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    print(f"  Accuracy: {acc:.4f}\n")

print("各FoldのAccuracy:", np.round(accuracies, 4))
print("平均Accuracy:", np.round(np.mean(accuracies), 4))
```
